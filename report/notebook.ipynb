```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker  # <-- nieuw import voor formatter

colors=['#1f78b4', '#e6550d']

rays = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 8192*2, 8192*4]
fpsGpu = [126.557287, 130.467026, 124.649205, 123.674965, 123.674965, 129.775327, 128.747240, 127.092551, 124.286413, 115.921696, 110.121171, 104.252, 93.252, 87.587463, 72.4118]
fpsCpu = [24.626318, 23.484180, 25.443637, 25.580356, 24.466385, 23.784696, 24.280541, 22.36, 20.9542, 18.252, 17.415464, 15.723720, 11.075, 7.39141, 4.25572]

plt.figure(figsize=(10, 6))

plt.plot(rays, fpsGpu, marker='o', linestyle='-', color=colors[0], label='GPU')
plt.plot(rays, fpsCpu, marker='o', linestyle='-', color=colors[1], label='CPU')

plt.title("FPS vs Rays")
plt.legend()
plt.xlabel("Rays")
plt.ylabel("FPS")
plt.xscale("log")
#plt.yscale("log")

plt.xticks(rays, rays)

# Zet de y-as formatter zodat gewone getallen worden gebruikt
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.ScalarFormatter())
ax.yaxis.set_minor_formatter(ticker.ScalarFormatter())

plt.grid(True, which="both", ls="--", lw=0.5)

plt.savefig("fps.png", format="png", bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

colors=['#1f78b4', '#e6550d']

# Aantal pixels als resolutie (bijv. ray count == pixel count)
resolutions = ["640x480", "800x600", "1280x720", "1920×1080" ]
fpsGpu = [327.082549, 235.42, 139.470052,87.783092 ]
fpsCpu = [48.49 ,24.802287 ,24.626318, 13.488]

# Posities op de x-as
x = np.arange(len(resolutions))
width = 0.35  # Breedte van de balken

plt.figure(figsize=(10, 6))

# Staafdiagrammen
plt.bar(x - width/2, fpsGpu, width, label='GPU', color=colors[0])
plt.bar(x + width/2, fpsCpu, width, label='CPU', color=colors[1])

# Labels en titel
plt.xlabel('Resolutie')
plt.ylabel('FPS')
plt.title('FPS per Resolutie (GPU vs CPU)')
plt.xticks(x, resolutions, rotation=45)
plt.legend()

# Grid en layout
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig("fps_vs_resolution_bar.png", format="png", bbox_inches='tight')
plt.show()
```
